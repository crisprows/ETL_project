```python
from bs4 import BeautifulSoup as bs
import requests
import os
```

```python
url="https://www.petsmart.com/"
response = requests.get(url)
```

```python
soup = bs(response.text, 'html.parser')
soup_meta = soup.find("meta", attrs={"name":"keywords"})
print(soup_meta)  
```

```python
keywords=soup_meta.get('content')
print(keywords)  
```

```python
keywords_list=keywords.split(",")
del keywords_list[0]
print(keywords_list)
    
```

```python
petsmart_keyword_dicts = []
key_list = ["meta_keyword"] 
  
# loop to iterate through elements 
# using dictionary comprehension 
# for dictionary construction 
n = len(keywords_list) 

for x in range(0, n): 
    petsmart_keyword_dicts.append({key_list[0]: keywords_list[x]}) 
    

    
print(petsmart_keyword_dicts)
```

```python
import pymongo
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.website_keywords
collection = db.petsmart
db.petsmart.delete_many({})

# figure out how to upload dictionary into the collection

collection.insert_many(petsmart_keyword_dicts)
```

```python
petsmart_keywords = db.petsmart.find()
for x in petsmart_keywords:
    print(x)
```
