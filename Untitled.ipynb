```python
from bs4 import BeautifulSoup as bs
import requests
import os
```

```python
url="https://www.petco.com/"
response = requests.get(url)
```

```python
soup = bs(response.text, 'html.parser')
soup_meta = soup.find("meta", attrs={"name":"keywords"})
print(soup_meta)  
```

```python
keywords=soup_meta.get('content')
print(keywords)
```
